```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

import seaborn as sns
```

```python
area_measurements = np.genfromtxt("CSVs/2nd area measurements.csv", delimiter=",")
pressure_measurements = pd.read_csv("CSVs/dady.csv")
```

```python
areas = area_measurements[1]
for i in range(len(pressure_measurements['Time (s) Run #1'])):
    try:
        areas[i]
    except:
        areas = np.append(areas, 0)
```

```python
pressure_measurements['Area [px^2]'] = areas
pressure_measurements.head(3)
```

```python
background = 'white'

if background == "black":
        plt.rcParams.update(
            {
                "lines.color": "w",
                "patch.edgecolor": "w",
                "text.color": "w",
                "axes.facecolor": background,
                "axes.edgecolor": "white",
                "axes.labelcolor": "w",
                "xtick.color": "w",
                "ytick.color": "w",
                "figure.facecolor": background,
                "figure.edgecolor": background,
                "savefig.facecolor": background,
                "savefig.edgecolor": background,
                "grid.color": "w",
                "axes.grid": False,
                'mathtext.fontset': 'dejavuserif',
                'font.family': 'serif'
            }
        )
else:
    plt.rcParams.update(
        {
            "lines.color": "k",
            "patch.edgecolor": "k",
            "text.color": "k",
            "axes.facecolor": background,
            "axes.edgecolor": "black",
            "axes.labelcolor": "k",
            "xtick.color": "k",
            "ytick.color": "k",
            "figure.facecolor": background,
            "figure.edgecolor": background,
            "savefig.facecolor": background,
            "savefig.edgecolor": background,
            "grid.color": "k",
            "axes.grid": False,
            'mathtext.fontset': 'dejavuserif',
            'font.family': 'serif'
        }
    )
```

```python
# setting color map
cmap = sns.color_palette("flare_r", as_cmap=True)

# preparing plot
fig, ax = plt.subplots(1,1, figsize=(10,6))

# setting up grid
ax.grid(True, which='major', alpha=.2)
ax.grid(True, which='minor', alpha=.1, linestyle=':')
ax.minorticks_on()

# plotting data
im = ax.scatter(c = pressure_measurements['Area [px^2]']/1e6, 
                y = pressure_measurements['Differential Pressure₁₋₂ (kPa) Run #1'],  
                x = pressure_measurements['Time (s) Run #1'], 
                marker = '.', 
                s = 20,
                cmap = cmap)

# colorbar and labels
fig.colorbar(im, label=r'Area [$kpx^2$]')
ax.set_xlabel(r'Time [$s$]')
ax.set_ylabel(r'Pressure [$kPa$]')

plt.savefig("Figures/Pressure vs Time/2nd run Pressure vs Time, with Area (in kpx^2).png", bbox_inches='tight', dpi=300)
plt.savefig("Figures/Pressure vs Time/2nd run Pressure vs Time, with Area (in kpx^2).svg", bbox_inches='tight')

plt.show()
```
